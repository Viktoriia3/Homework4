```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
import matplotlib as plt
import matplotlib as mpl
import matplotlib.cm as cm 
import matplotlib.pyplot as plt
from matplotlib import pyplot

olympic_df = pd.read_csv(r'D:/athlete_events.csv')
olympic_df.head(2)
```

Пропущенные значения есть в столбцах: возраст, вес, рост. И медали у тех, кто их не завоевал

Посмотрим распределение спортсменов по возрасту. Больше всего спортсменов-участников олимпиад в возрасте от 20 до 23 лет.

```python
plt.hist(olympic_df['Age'], color = 'blue', edgecolor = 'black',
         bins = int(180/5))
sns.distplot(olympic_df['Age'], hist=True, kde=False, 
             bins=int(180/5), color = 'blue',
             hist_kws={'edgecolor':'black'})
plt.title('Histogram of Age')
plt.ylabel('Membors')
```

```python
#замена данных в столбце медали, золото - 1, серебро - 2, бронза - 3
olympic_df = olympic_df.replace(['Gold', 'Silver', 'Bronze'], ['1', '2', '3'])
print(olympic_df)
```

```python
olympic_df.dropna (subset=['Medal']) #оставим только спортсменов, которые завоевали медали
```

```python
sns.countplot(x='Sex',data=olympic_df) 
plt.title('Распределение медалистов по полу')
```

Среди медалистов олимпиад мужчин 2,5 раза больше

```python
sns.countplot(x='Season',data=olympic_df) 
plt.title('Количество разыгранных наград на зимних и летних ОИ')
```

На летних ОИ разыгрывается гораздо больше медалей 

```python
goldMedalsall = olympic_df[(olympic_df.Medal == '1')] #Только золотые медалисты
```

```python
totalGold = goldMedalsall.Team.value_counts().reset_index(name='Medal').head(5)
g = sns.catplot(x="Team", y="Medal", data=totalGold,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Страны")
g.set_ylabels("Количество медалец")
plt.title('Топ 5 стран по количеству медалей')
plt.show()
```

Топ 5 стран по количеству медалей за историю ОИ: США, СССР, Германия, Италия и Великобритания

```python
#Оставляем только Норвегию для анализа
Norway_df = olympic_df[olympic_df['Team'] == 'Norway']
print(Norway_df)
```

```python
Norway_df.dropna (subset=['Medal'])
```

```python
medal_totals = Norway_df.groupby("Medal")["Year"].sum().sort_values()
```

```python
medal_totals.plot(kind="barh", fontsize=5)
```

```python
#золотые медалисты Норвегии
goldMedals = Norway_df[(Norway_df.Medal == '1')]
print(goldMedals.head())
```

```python
#золотые медалисты Норвегии по возрасту
plt.figure(figsize=(20, 10))
plt.title('Distribution of Gold Medals')
sns.countplot(goldMedals['Age'])
plt.show()
```

```python
#Интересно посмотреть в каком виде спорта победили те, кому больше 40 лет
goldMedals = Norway_df[(Norway_df.Medal == '1')]
print('The no of athletes is',
      goldMedals['ID'][goldMedals['Age'] > 40].count(), '\n')
print(goldMedals[goldMedals['Age'] > 40])
```

```python
athlets = goldMedals['Sport'][goldMedals['Age'] > 40]
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(athlets)
plt.title('Золотые медалисты возраста больше 40 лет')
plt.show()
```

Это были все спорстмены по стрельбе :)

```python
#количество мужчин в летних играх
menOlympics = Norway_df[(Norway_df.Sex == 'M') &
                         (Norway_df.Season == 'Summer')]
print(menOlympics.head(10))
 
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=menOlympics)
plt.title('Медалисты мужчины в летних играх')
plt.show()
```

```python
#количество мужчин в зимних играх
menOlympics = Norway_df[(Norway_df.Sex == 'M') &
                         (Norway_df.Season == 'Winter')]
print(menOlympics.head(10))
 
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=menOlympics)
plt.title('Медалисты мужчины в зимних играх')
plt.show()
```
